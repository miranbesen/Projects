```python
import random
import string
from enum import Enum
import sys
```

```python
class Orientation(Enum): #enum typemiz, degerleri istenilen sekle gore atadık. 
    soldanSaga= 0
    sagdanSola= 1
    asagidanYukari= 2
    yukaridanAsagi= 3
   
```

```python
def createSut(n):#sutunları random harflere gore olusturuyrouz
    sutList=[]
    if(n<0):
        print("Böyle bir eleman sayisi ile vektör olusturamayız.")
        return 
    if(n==0): #sutun sayisi 0 ise direk 0'ı yolluyoruz.
        return [0]
    for i in range(n): #sutun sayisi kadar random harf olustuyoruz ve bunları sutun listemize ekliyoruz
        randomLetter=random.choice(string.ascii_lowercase)
        sutList.append(randomLetter)
    return sutList
```

```python
def createLetterMat(m,n): #harflerden olusan matrisimiz icin yazdıgımız fonksiyon
    if(m<=0): 
        print("Olusturmak istediginiz matrisin satir sayisi uyusmuyor.")
        return [0]
    if(n<=0):
        print("Olusturmak istediginiz matrisin sutun sayisi uyusmuyor.")
        return [0]
    m1=[]
    for i in range(m): #satir sayimiz kadar döner. Sütunu oluşturmak icin sutun sayisini createSut fonksiyonumuza yollar.
        m1.append(createSut(n)) #listemize ekliyoruz.
        
    return m1
```

```python
def newMatWord(m1,n1,myword,orientation): #Bu fonksiyonumuzun amaci elimizde olan matrisimize istedigimiz yerden itibaren istedigimiz kelimeyi yazdırmak.
    
    if(m1>m): #eger matrisimizin satır sayısından büyük bir degerden baslamak isterse, yazdıramıcagımız icin cıkıyoruz.
        sys.exit('satir sayisini fazla girdin')
    if(m1==0): #matrisimiz 0 satırdan baslayamaz.
        sys.exit('0 girilmez satira')
    if (n1>n): #eger matrisimizin sutun sayısından büyük bir degerden baslamak isterse, yazdıramıcagımız icin cıkıyoruz.
        sys.exit('sutun degerini fazla girdin')
    if (n1==0):#matrisimiz 0 sutundan baslayamaz.
        sys.exit('0 girilmez sutuna')
    
    kelimeUzunluk=len(myword) #yazacagımız kelimenin uzunlugu kadar döngümüz dönecek.
    for i in range(kelimeUzunluk):     
        if(orientation==Orientation.soldanSaga.value): #eğer orientation degerimiz 0 ise soldanSaga yazıcak 
            createMat[m1-1][n1-1]=myword[i] #matrisimizde istedigimiz yerden itibaren  soldan saga dogru elemanları eklemeye baslıyoruz. 
            n1+=1 #sutun sayisini arttırdık.(Cünkü soldan saga dogru yazıyoruz)
            if(n1>n):#eger sutun sayisi toplam sutun sayimizi gecti ise, bir alt satırın 1 sütünündan devam edicez.
                n1=1 
                m1+=1 #satır degerimizi arttırdık.
                if(m1>m): #eger satır degerimiz matrisimizin toplam satırından büyük olursak tekrar ilk satıra dön.
                    m1=1 #burda 1 e esitleme sebebimiz ise -1 cıkardıgımız icin normal olarak 0 a erismis oluyoruz kelimeyi yazarken matrisimize.
        
        if(orientation==Orientation.sagdanSola.value): #eğer orientation degerimiz 1 ise sagdanSola yazıcak 
            createMat[m1-1][n1-1]=myword[i] #matrisimizde istedigimiz yerden itibaren  sagdan sola dogru elemanları eklemeye baslıyoruz. 
            n1-=1 #sutun sayısını bu sefer 1 azaltacaz ki sola dogru hareket etmemizi saglasın
            if(n1==0): #eger sutun sayimiz 0 olursa sutunu yeniden toplam sutun uzunlugumuzdan baslatıyoruz ve satırımızı 1 azaltıyoruz.
                n1=n
                m1-=1
                if(m1==0): #eger satır sayımız minimuma indi ise tekrar matrisimizin satır sayısından devam edicez.
                      m1=m      
                  
        if(orientation==Orientation.asagidanYukari.value): #eğer orientation degerimiz 2 ise asagidanYukari yazıcak 
            createMat[m1-1][n1-1]=myword[i] #matrisimizde istedigimiz yerden itibaren asagıdan yukarı dogru elemanları eklemeye baslıyoruz. 
            m1-=1 #satırı 1 azaltarak asagıdan yukarı dogru cıkmamızı saglar.
            if(m1==0): #eger satır sayımız minimuma gelirse, satırımızı yeniden toplam satır uzunlugumuzdan baslatıyoruz ve sutun sayımızı bir azaltıyoruz.
                m1=m 
                n1-=1 #asagıdan yukarı yazarken sola dogru yani yukarıdan asagı yazarken saga dogru hareket etmesinin tersine dogru hareket edecegi icin bundada 1 cıkartarak sola dogru yani ters yonde hareket etmemizi saglıyor.
                if(n1==0): #sutun sayımızı minimuma ulastıgı zaman sutunumuzu tekrar matrisimizin sutun sayımızdan baslatıyoruz
                      n1=n
        
        if(orientation==Orientation.yukaridanAsagi.value): #eğer orientation degerimiz 3 ise yukaridanAsagi yazıcak 
            createMat[m1-1][n1-1]=myword[i] #matrisimizde istedigimiz yerden itibaren yukarıdan asagı dogru elemanları eklemeye baslıyoruz. 
            m1+=1 #yukarıdan asagı dogru yazacagımız icin satır sayımızı 1 arttırıyoruz.
            if(m1>m): #satır sayımız matrisimizin satır sayısını gecer ise satır sayımızı tekrar matrisimizin minimum satır sayısına getiriyoruz ve sutun sayımızı 1 arttırarak bir sonraki sutundan itibaren devam ediyoruz.
                m1=1 
                n1+=1
                #ekstra olarak >n deme sebebimiz >=n demek yerine, 1 den baslattıgımız icin sutun ve satır sayımızı normalden 1 ekstra büyük oldugu icin >n dememiz gerekir.
                if(n1>n): #eger sutun sayımız matrisimizin sutun sayısını gecer ise sutun sayımızı tekrar matrisimizin sutun sayısından baslatıyoruz.
                      n1=1    
```

```python
def Palindrome(kelime):   
    for i in range(0, int(len(kelime)/2)):  #i degerimiz 0'dan kelimenin yarısına kadar dönecek, çünkü bas ve son kelimeyi kıyaslayacak, her seferinde 2 kelimeyi kıyaslayacagı icin yarısına kadar dönmesi yeterli
        
        if kelime[i] != kelime[len(kelime)-i-1]: #eger kelimelerden herhangi biri aynı olmazsa palindrome olmuyor, yani false döndürüyoruz. 
            return False
        
    return True
```

```python
def palindromMatWord(createMat,orientation): #matrisimizde bulunan palindrome kelimeleri kullanıcının verdigi yön degerine göre buluyor.
    polindromList=[]
    denemeString=''
    satS=len(createMat) #satır sayımız matrisimizdeki
    sutS=len(createMat[0]) #sutun sayımız (listemizdeki 0 elemanda kac deger olduguna gore belirleyebiliriz.) 
    publicLabel=satS*sutS #toplam döngünün dönmesi gereken uzunluk
    m0=1 #m0 satırı temsil ediyor, 1 den baslama sebebi matrisimizin ilk satırından baslatıyoruz. Bunu kullandıgımız orientationlar ise soldanSaga ve yukarıdan asagı olursa.(cunku bunlar duzgun bir sekilde devam eder.)
    n0=1#n0 sutun temsil ediyor, 1 den baslama sebebi matrisimizin ilk sutundan baslatıyoruz. Bunu kullandıgımız orientationlar ise soldanSaga ve yukarıdan asagı olursa.(cunku bunlar duzgun bir sekilde devam eder.)
    m1=satS #m1 satırı temsil ediyor, matrisimizin satır sayısına esit cunku sagdan sola kelimelerin palindromesine bakarken ya da asagıdan yukarı olan palindromelere bakarken matrisin son satırından  baslayarak kontrol etmeye baslıyoruz cunku ters olur
    n1=sutS #m1 sutun temsil ediyor, matrisimizin sutun sayısına esit cunku sagdan sola kelimelerin palindromesine bakarken ya da asagıdan yukarı olan palindromelere bakarken  matrisin son sutunundan baslayarak kontrol etmeye baslıyoruz cunku ters olur
    if(orientation==Orientation.soldanSaga.value):  #eğer orientation degerimiz 0 ise soldan saga dogru bakıcak
        for i in range(publicLabel): #burda amacımız matristeki tüm elemanları kendinden itibaren matrisimizdeki son elemana kadar palindromelerine kontrol ederek palindrome olanları bir listede toplamak onun icin toplam matrisimizdeki eleman sayısı kadar donmus olacak.
            denemeString='' #burda her seferinde aynı kelimelerin uzerine eklemesin diye eger matrisimizdeki bir eleman icin kontrol yapıldıysa bu sefer diger eleman icin kontrol yapılması acısından bos bir stringe donustururuz.       
            for j in range(publicLabel): #bunu da toplam publicLabel kadar dondurme sebebimiz matrisimizde bulunan son sutun ve satırdaki elemanımız da son kez tum elemanları kıyaslama yaparak palindrome olup olmadıgına bakıyoruz.
                   denemeString+=createMat[m0-1][n0-1] #newmatWord fonksiyonumuzda yaptıgımız soldan saga kelime yazma yerine o kelimeleri stringte tutmayı sagladık.Tek farkı bu.
                   n0+=1 #yine diger fonksiyonumuzdaki gibi soldan saga dogru gidecegimiz icin sutun sayısını bir arttırdık.
                   if(n0>sutS):
                        n0=1
                        m0+=1
                        if(m0>satS):
                            m0=1
                   #print(denemeString)
                   stringLen=len(denemeString) #kelime ekledigimiz stringin uzunlugunu buluyoruz.
                   #print(n)
                   if(stringLen>=3): #eger ekledigimiz kelimelerin uzunlugu 3 ten buyuk ise ife girer
                       bayrak=Palindrome(denemeString) #eger kelimemiz palindrome ise true döneri degilse false döner.
                       #print(bayrak)
                       if bayrak: #eger true ise yani kelime palindrome ise palindrome listemize stringimizi ekliyoruz.
                           polindromList.append(denemeString)
                           #print(polindromList)
            n0+=1 #burda sutun degerimizi 1 arttırma sebebimiz ise matrisimizde bulunan  ilk sutun ve satırdaki eleman icin tum kontrolleri sagladıktan, sonra ilk satır ve 2 sutundaki eleman icin palindrome kontrolu yapmak bu sekilde tum sutunları bitirdikten sonra bir sonraki satırdaki ilk sutundan baslayarak yine tum sutundaki elemanları kontrol ederek,bir alt satıra gecer boylece matriste bulunan tüm elemanlar icin kontrol saglanmıs olur.
            if(n0>sutS):
                   n0=1
                   m0+=1
                   if(m0>satS):
                        m0=1
  
    if(orientation==Orientation.sagdanSola.value): #eğer orientation degerimiz 1 ise sagdan sola dogru bakıcak
        #yine bundada yukarıda yazdıgımız gibi mantık aynı fakat tek farkı bundan sagdan sola baktıgı icin matrisimizde ki elemanları sagdan sola bir sekilde stringimize ekleyerek kontrolleri yapar.
        #burada n1 ve m1 matris degerleri kullanılıyor ccunku sagdan sola kontrol icin matrisimizin son satır ve sutundan itibaren stringimize ekleyerek gider.(Ters seklinde düsünecez soldan sagadan)
        for i in range(publicLabel):
            denemeString=''        
            for j in range(publicLabel):
                denemeString+=createMat[m1-1][n1-1]
                n1-=1
                if(n1==0):
                    n1=n
                    m1-=1
                    if(m1==0):
                          m1=m
                
                #print(denemeString)
                stringLen=len(denemeString)
                   #print(n)
                if(stringLen>=3):
                     bayrak=Palindrome(denemeString)
                     #print(bayrak)
                     if bayrak:
                        polindromList.append(denemeString)
                          #print(polindromList)

            n1-=1 #burda ise sagdan sola kontrol yaptıgımız icin sutun degerini bir azaltarak bir sonraki sutun degerinden itibaren tüm matristeki elemanları sagdan sola dogru kontrol etmemizi saglıcak.
            if(n1==0):
                 n1=n
                 m1-=1
                 if(m1==0):
                      m1=m
            
    if(orientation==Orientation.asagidanYukari.value):  #eğer orientation degerimiz 2 ise asagıdan yukarı dogru bakıcak
        #yine bundada yukarıda yazdıgımız gibi mantık aynı fakat tek farkı bundan asagıdan yukarı baktıgımız icin matrisimizde ki elemanları asagıdan yukarı bir sekilde stringimize ekleyerek kontrolleri yapar.
        #burada n1 ve m1 matris degerleri kullanılıyor cunku asagidan yukarı kontrol icin matrisimizin son satır ve sutundan itibaren stringimize ekleyerek gider.(Ters seklinde düsünecez yukarıdan asagıdan)
        for i in range(publicLabel):
            denemeString=''        
            for j in range(publicLabel):
               denemeString+=createMat[m1-1][n1-1] 
               m1-=1
               if(m1==0):
                    m1=m
                    n1-=1
                    if(n1==0):
                          n1=n
               #print(denemeString)
               stringLen=len(denemeString)
                   #print(n)
               if(stringLen>=3):
                     bayrak=Palindrome(denemeString)
                     #print(bayrak)
                     if bayrak:
                        polindromList.append(denemeString)
                          #print(polindromList)            


            m1-=1#burda ise asagidan yukarı kontrol yaptıgımız icin satir degerini bir azaltarak bir sonraki satir degerinden itibaren tüm matristeki elemanları asagidan yukarı dogru kontrol etmemizi saglıcak.
            if(m1==0):#eger satır sayımız minimuma gelirse, satırımızı yeniden toplam satır uzunlugumuzdan baslatıyoruz ve sutun sayımızı bir azaltıyoruz.
                m1=m
                n1-=1#asagıdan yukarı yazarken sola dogru yani yukarıdan asagı yazarken saga dogru hareket etmesinin tersine dogru hareket edecegi icin bundada 1 cıkartarak sola dogru yani ters yonde hareket etmemizi saglıyor.
                if(n1==0): 
                    n1=n

    if(orientation==Orientation.yukaridanAsagi.value):#eğer orientation degerimiz 3 ise yukaridan asagı dogru bakıcak
         #yine bundada yukarıda yazdıgımız gibi mantık aynı fakat tek farkı bundan yukarıdan asagı bir sekilde baktıgımız icin matrisimizde ki elemanları yukarıdan asagi bir sekilde stringimize ekleyerek kontrolleri yapar.
         #burada m0 ve n0 dan baslama sebebimiz ise yukarıdan asagı dogru kontrol yaparken ilk satırın ilk sutunundan asagı dogru kontrol yapmamız.
            for i in range(publicLabel):
                denemeString=''        
                for j in range(publicLabel):
                    denemeString+=createMat[m0-1][n0-1]
                    m0+=1
                    if(m0>m):
                        m0=1
                        n0+=1
                        if(n0>n):
                              n0=1                             
                    stringLen=len(denemeString)
                    #print(denemeString)   
                    #print(n)
                    if(stringLen>=3):
                         bayrak=Palindrome(denemeString)
                         #print(bayrak)
                         if bayrak:
                            polindromList.append(denemeString)
                              #print(polindromList)                                 
                m0+=1 #yukarıdan asagı kontrol yaptıgımız icin ilk satır ilk sutun icin kontrol yaptıktan sonra ikinci satır ilk sutundaki elemandan itibaren kontrol yapar sonra ucuncu satır ilk sutundaki elemandan sonra kontrol yapar bu sekil de dusunuyoruz.
                if(m0>m):
                     m0=1
                     n0+=1
                     if(n0>n):
                         n0=1   

    return polindromList  
```

```python
def printMat(createMat,m,n): #matrisimizi ekrana bastırıyoruz.
    for i in range(m):
        for j in range(n):
            print(createMat[i][j],end="  ")
        print()
```
